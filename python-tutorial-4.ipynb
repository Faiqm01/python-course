## Functions

As well as loading modules, we can make our own functions and modules.
(almost) every module in Python is just a series of other python code,
so we can make our own to extend available functions.

To make our own functions we just need to use the word 'def'

```
def my_func():
    pass
```

Above we've made our own function, like other python functions we use ()
to illustrate its a function. Currently the only thing it does is 'pass'
which means 'pass by' i.e. ignore. Like loops, functions are grouped by
indentations, so everything indented is in our function.

```
def helloWorld():
    print("Hello world")
```

Now we've made another function, 'helloWorld'. When we run this above
cell, nothing happens, thats because we've defined our function, but not
ran it.

```
helloWorld()
```

Now we run our function just like any function. We can pass data to our
functions just like we do with any other function.

```
def addingFunc(a, b):
    print(a+b)

addingFunc(1, 20)
```

We refer to these as "arguments", so the function "addingFunc" takes 2
arguments, "a" and "b"

### Optional arguments

In addition to normal arguments, we can also use "optional arguments"

```
def mathsFunc(a, b, adding=False, product=False):
    output = 0
    if adding:
        output += a + b
    if product:
        if output == 0:
            output += 1
        output *= a * b
    print(output)

mathsFunc(10, 2, adding=True)
mathsFunc(10, 2, product=True)

```

Functions are helpful because a bit like for loops, it helps us tidy our
code up, and make it reproducable

### Type hints

Sometimes functions will give you a 'type hint' which gives you
information on the type of data it expects to receive and which it will
output

```


def mathsFunc(a:int, b:int, adding:bool=False, product:bool=False) -> None:
    output = 0
    if adding:
        output += a + b
    if product:
        if output == 0:
            output += 1
        output *= a * b
    print(output)

mathsFunc(10, 2, adding=True)
mathsFunc(10, 2, product=True)

```

So our above function is identical, but now we've let other people know
that a and b are supposed to be ints and that adding and product are
bools.

### DocStrings

As well as type hints, functions also often have something called a 'doc
string' which is a piece of text describing the function. For small
functions they're often a sentence, but for bigger ones they can really
help explain what you're doing

```

def mathsFunc(a, b, adding=False, product=False):
    """
    Function to add or multiple two integers together
    """
    output = 0
    if adding:
        output += a + b
    if product:
        if output == 0:
            output += 1
        output *= a * b
    print(output)

mathsFunc(10, 2, adding=True)
mathsFunc(10, 2, product=True)

```
